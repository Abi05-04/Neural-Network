```python
from sklearn.datasets import load_digits
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

# 🔹 Load dataset
data = load_digits()
X = data.data          # (1797, 64)
y = data.target        # (1797,)

# 🔹 Convert to tensors
X = torch.FloatTensor(X)
y = torch.LongTensor(y)

# 🔹 Create Dataset and DataLoader (BATCHES)
dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

# 🔹 Define model
model = nn.Sequential(
    nn.Linear(64, 32),
    nn.ReLU(),
    nn.Linear(32, 10)   # 10 classes (0–9)
)

# 🔹 Loss function and optimizer
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 🔹 Number of epochs (more loops)
epochs = 100

# 🔹 Training loop
for epoch in range(epochs):
    total_loss = 0

    for xb, yb in loader:   # ✅ batches used here
        optimizer.zero_grad()

        output = model(xb)        # logits
        loss = loss_fn(output, yb)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss:.4f}")

# 🔹 Evaluate accuracy
correct = 0
total = 0

with torch.no_grad():
    for xb, yb in loader:
        output = model(xb)
        preds = torch.argmax(output, dim=1)

        correct += (preds == yb).sum().item()
        total += yb.size(0)

print("Accuracy:", correct / total)
```

```python
# 🔹 1. Imports
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.datasets import load_digits
import numpy as np
import cv2
from google.colab import files

# 🔹 2. Load dataset
data = load_digits()
X = data.data        # (1797, 64)
y = data.target      # (1797,)

# 🔹 3. Convert to tensors
X = torch.FloatTensor(X)
y = torch.LongTensor(y)

# 🔹 4. Create batches
dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

# 🔹 5. Define model
model = nn.Sequential(
    nn.Linear(64, 32),
    nn.ReLU(),
    nn.Linear(32, 10)
)

# 🔹 6. Loss and optimizer
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 🔹 7. Training loop (with batches)
epochs = 50

for epoch in range(epochs):
    total_loss = 0

    for xb, yb in loader:   # 🔥 batches used here
        optimizer.zero_grad()

        output = model(xb)
        loss = loss_fn(output, yb)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss:.4f}")

# 🔹 8. Create and save a sample image (digit-like)
sample = data.images[0]     # 8x8 image
sample = (sample * 16).astype("uint8")  # scale to save

cv2.imwrite("three.png", sample)

# 🔹 Show saved file in Colab
import os
print("Files:", os.listdir())

# 🔹 9. Load and preprocess image
img = cv2.imread("three.png")

grar = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rszar = cv2.resize(grar, (8, 8), interpolation=cv2.INTER_AREA)

# Normalize like training
rszar = rszar / 16.0

# Flatten
far = rszar.reshape(1, 64)

# Convert to tensor
fimg = torch.FloatTensor(far)

# 🔹 10. Prediction
output = model(fimg)
pred = torch.argmax(output, dim=1)

print("Prediction:", pred)   # 👉 tensor([1]) (example)

# 🔹 11. Optional: download image
files.download("three.png")
```
